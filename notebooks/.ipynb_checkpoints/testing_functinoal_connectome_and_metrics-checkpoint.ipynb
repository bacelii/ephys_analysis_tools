# Step 1. Load the data

```python
from ephys_analysis_tools import data_api as api
data = api.example_data_load()
```

# Step 2. Compute the functional connectome

```python
connectivity_method = "pearson"
```

```python
conn_result = conn.compute_connectivity(
    data,
    method = connectivity_method
)
conn_result
```

```python
graph = g.graph_from_connectivity_result(
    conn_result,
    threshold = 0.4,
    #density=0.15,
)
graph
```

```python
metrics = m.compute_graph_metrics(graph)
metrics
```
